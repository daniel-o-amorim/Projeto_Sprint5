```python
#Importações de bibliotecas:

import pandas as pd
import plotly_express as px
import streamlit as st
```

```python
#Acesso e leitura do CSV:

df_vehicles = pd.read_csv("../vehicles_us.csv")
print("Informações do DataFrame:")
print()
print(df_vehicles.info())

print(" Cinco primeiras linhas do DataFrame:")
print()
print(df_vehicles.head())
```

```python
#Ver total de valores nulos por coluna:
df_vehicles.isna().sum()
```

```python
#Limpar valores nulos:
df_vehicles_us = df_vehicles.dropna()
print("Total de valores nulos por coluna após limpeza:")
df_vehicles_us.isna().sum()
```

```python
#Coluna Price:

#Ordenar valores da coluna:
df_vehicles_us_sorted = df_vehicles_us.sort_values(by="price")
#Nova clonuna com valores em USD:
df_vehicles_us_sorted["price_usd"] = df_vehicles_us_sorted["price"].apply(lambda x: f"${x:,.2f}")

#Gráfico de Dispersão:
fig = px.scatter(
    df_vehicles_us_sorted,
    x=df_vehicles_us_sorted.index,
    y='price',
    title='Dispersão dos Preços dos Veículos',
    labels={'price': 'Preço (USD)', 'index': 'Índice do DataFrame'},
    hover_name='price_usd'
    )
#Exibir gráfico: 
fig.show()
```

```python
#Histogramas:
# Distribuição de Preços dos Veículos:
fig_price = px.histogram(
    df_vehicles_us,
    x='price',
    title='Distribuição de Preços dos Veículos',
    labels={'price': 'Preço (USD)'},
    nbins=50,
)
fig_price.show()
# Distribuição de Quilometragem dos Veículos:
fig_odometer = px.histogram(
    df_vehicles_us,
    x='odometer',
    title='Distribuição de Quilometragem dos Veículos',
    labels={'odometer': 'Quilometragem (milhas)'},
    nbins=50,
)
fig_odometer.show()

#Distribuição de Quilometragem dos Veículos (em km):
#Nova coluna com valores em km:
df_vehicles_us.loc[:, 'odometer_km'] = df_vehicles_us['odometer'] * 1.60934

#Gráfico de Dispersão:
fig_odometer_km = px.histogram(
    df_vehicles_us,
    x='odometer_km',
    title='Distribuição de Quilometragem dos Veículos (em km)',
    labels={'odometer_km': 'Quilometragem (km)'},
    nbins=50,
)
fig_odometer_km.show()
```
